# Unstack and stack

```python
df = pd.read_csv('C:/Personel/LearnPython/code/drinks.csv')
df.set_index(['continent','country'], inplace=True)
```

```python
df
```

```python
stack_df = df.stack() # All columns converted to rows
```

```python
stack_df
```

```python
stack_df.unstack() # All rows converted to columns
```

```python
stack_df.unstack().sort_index()
```

```python
stack_df
```

```python
stack_df.unstack('country')
```

```python
stack_df
```

```python
stack_df.unstack(2) # 3rd level is beer_servings, spirit_servings etc
```

```python
df
```

```python
test_df = df.set_index('beer_servings', append=True)
```

```python
test_df
```

```python
test_df.stack().unstack().sort_index()
```
