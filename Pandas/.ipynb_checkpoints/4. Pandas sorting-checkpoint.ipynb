```python
import pandas as pd
```

```python
names = ["age","workclass","fnlwgt","education","edu-num","marital-status","occupation","relationship","race","sex","cap-gain","cap-loss","hour/week","country","salary"]
df = pd.read_csv('C:/Personel/LearnPython/code/adult.data.csv', header=None, names=names)
df.head(3)
```

```python
df['country'].nunique()
```

```python
# sort values based on column name
df.sort_values(by=['country']) # ascending 
df.sort_values(by=['country'], ascending=False) # descending
```

```python
df.sort_values(by=['age']) # ascending
df.sort_values(by=['age'], ascending=False) # descending
```

## multiple columns to sort, it will sort first first column and within that it will sort for second column

```python
# multiple columns to sort, it will sort first by country and in each country by age

df.sort_values(by=['country','age'])
```

```python
df.sort_values(by=['country','age'], ascending=False)
```

### Lets say we want to make country descending and age as ascending

```python
df.sort_values(by=['country','age'], ascending=[False, True])
```

# Sort index and inplace parameter exists in sort_values()

```python
df.head(2)
```

```python
df.sort_values(by=['country','age'], ascending=[False, True])
```

```python
df.sort_values(by=['country','age'], ascending=[True, True], inplace=True)
```

```python
df.head()
```

```python
# sort index can be in ascending and descending in nature
df.sort_index()
```
