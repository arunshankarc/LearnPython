# Broadcasting Operations

### Run one operation against all the rows of a given column

```python
# For example let me add 5 to all age
# and each fnlwgt by 1.2
# and each cap-gain by 0.2 since age has been incremented by 5
# One operation broadcasted all the changes for a given rows for a column
df['age'] = df['age'] + 5
df['fnlwgt'] = df['fnlwgt'] * 1.2
df['cap-gain'] = df['cap-gain'] + (df['cap-gain'] * .2)
df.head(2)
```

```python
list(df.columns) # convert the names of columns to list which helps in processing wherever needed
```
