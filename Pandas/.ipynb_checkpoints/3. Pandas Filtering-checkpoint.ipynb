import pandas as pd
import numpy as np

# Filter data

```python
names = ["age","workclass","fnlwgt","education","edu-num","marital-status","occupation","relationship","race","sex","cap-gain","cap-loss","hour/week","country","salary"]
df = pd.read_csv('C:/Personel/LearnPython/code/adult.data.csv', header=None, names=names)
df
```

```python
# sex column has leading spaces
df['sex'] = df['sex'].apply(lambda x: x.strip())
df.head()
```

```python
# condition is a series object and when we do df[condition] we only get values which returned as True
condition = df['sex'] == 'Male'
df[condition]
```

```python
# Simple way is as below

male_data_df = df[df['sex'] == 'Male']
male_data_df.head()
```

```python
# Check age greater than 50 years

age_greater_than_50_years_df= df[df['age'] >= 50]
age_greater_than_50_years_df
```

# Merge multiple conditions and filter

```python
# Do not forget to add '()' for each conditions else it will fail with TypeError
male_age_greater_than_50_df = df[(df['sex'] == 'Male') & (df['age'] >= 50)]
male_age_greater_than_50_df
```

### Lets say I want to filter data in such a way that for a given column I need to check for multiple conditions, then in previous way I will have to add multiple conditions which is cumbersome, so in simple terms just use isin()
### For example education should be 10th, 11th, 12th, Bachelors, Masters

```python
df['education'] = df['education'].apply(lambda x: x.strip())
df[df['education'].isin(['10th', '11th', '12th', 'Bachelors', 'Master'])]
```

# between()

```python
age_between_59_60_years = df[df['age'].between(59, 60)]
```
