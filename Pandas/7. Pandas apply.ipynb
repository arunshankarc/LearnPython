```python
import pandas as pd
names = ["age","workclass","fnlwgt","education","edu-num","marital-status","occupation","relationship","race","sex","cap-gain","cap-loss","hour/week","country","salary"]
df = pd.read_csv('C:/Personel/LearnPython/code/adult.data.csv', header=None, names=names)
```

# .apply() on single column

```python
df.head(10)
```

```python
def convert_salary_to_easy_calc(salary):
    if salary == ' <=50K':
        return 1
    else:
        return 2
```

```python
df['changed_sal'] = df['salary'].apply(convert_salary_to_easy_calc)
df.head(10)
```

# .apply() on multiple columns

```python
df.head(10)
```

```python
# Where row is entire row
def description(row):
    return f"{row[9]} is earning {row[14]}"
```

```python
df['small_description'] = df.apply(func=description, axis='columns')
```
