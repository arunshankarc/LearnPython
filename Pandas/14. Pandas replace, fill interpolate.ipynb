```python
import pandas as pd
import numpy as np
df = pd.read_csv('input_files/weather_data.csv', parse_dates=['day']) # parse_dates converts to date column
type(df.day[0])
df
```

```python
with_index_df = df.set_index('day')
```

```python
with_index_df
```

# Replace NA with meaningful values

```python
new_df = with_index_df.fillna(0)
```

```python
new_df
```

```python
new_df = with_index_df.fillna({
    'temperature': 9999,
    'windspeed': 8888,
    'event': 'No Event'
})
```

```python
new_df
```

```python
with_index_df.fillna(method='ffill')  # Fill same as previous element i.e. same as previous day's value
```

```python
with_index_df.fillna(method='bfill') # Copy next day's value
```

```python
with_index_df.fillna(method='bfill', axis="rows") # Copy next day's value
```

```python
with_index_df.fillna(method='bfill', axis="columns") # Copy next day's value
```

```python
 with_index_df.interpolate() # Making better guess - linear interpolation
```

```python
 with_index_df.interpolate(method='time') # make guess based on previous and current date
```
