```python
import pandas as pd
df = pd.read_csv('C:/Personel/LearnPython/code/drinks.csv', index_col='country') # Set index which is different other than 0,1,2..
```

```python
df.head()
```

```python
df.loc['Albania'] # In simple terms this is to fetch the row based on index
```

```python
df.loc['India']
```

```python
df.loc[['India', 'Albania', 'Afghanistan']] # This is to get multiple indexes (rows)
```

```python
# Get specific columns for specific country
df.loc[['India', 'Albania', 'Afghanistan'], ['beer_servings', 'continent']] 
```

```python
# Throws KeyError for those indexes which does not exist
df.loc[['India', 'Albania', 'XYZ'], ['beer_servings', 'continent']] 
```

```python
df.head(10)
```

```python
df.loc['Argentina':'Australia'] # between 2 indexes
```

```python
# Get specific columns for the data between 2 indexes
df.loc['Argentina':'Australia', ['beer_servings', 'continent']]
```
