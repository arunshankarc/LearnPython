# .agg()

```python
df_country.agg('sum')
```

```python
df_country.agg('sum')['SALES']
```

```python
df_country.agg('sum')[['SALES','PRICEEACH']] # Multiple columns
```

```python
df_country.agg(['sum', 'mean'])[['SALES','PRICEEACH']] # Multiple columns and multiple functions
```

```python
df_country.agg(['sum', 'mean', 'min', 'max'])[['SALES','PRICEEACH','QUANTITYORDERED']] # Multiple columns and multiple functions
```

```python
df_country.get_group('Australia')['SALES'].min()
```

```python
df_country.agg({'QUANTITYORDERED':'mean','SALES':'sum'}) # apply specific function to specific columns
```
