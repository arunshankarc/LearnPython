# Pandas Options

### max_rows, max_columns

```python
import pandas as pd
import numpy as np
names = ["age","workclass","fnlwgt","education","edu-num","marital-status","occupation","relationship","race","sex","cap-gain","cap-loss","hour/week","country","salary"]
df = pd.read_csv('C:/Personel/LearnPython/code/adult.data.csv', header=None, names=names)
```

```python
pd.options.display.max_columns
```

```python
pd.options.display.max_rows
```

```python
pd.options.display.max_columns = 10
```

```python
pd.options.display.max_rows = 10
```

```python
df
```

```python
pd.get_option('max_row')
```

```python
pd.get_option('max_columns')
```

```python
pd.reset_option('max_row')
```

```python
pd.reset_option('max_columns')
```

```python
df
```

```python
pd.options.display.max_rows
```

```python
pd.set_option('max_rows', 6)
```

```python
df
```

```python
pd.describe_option('max_columns')
```
